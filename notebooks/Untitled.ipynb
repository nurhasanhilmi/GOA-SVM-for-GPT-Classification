```python
import numpy as np
import pandas as pd

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
data = pd.read_csv('../data/gpt_mfccs_stats.csv')
data
```

```python
x = data.iloc[:,65:91]
y = data['technique']
x
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf = SVC(kernel='rbf', decision_function_shape='ovo', random_state=42)
scores = cross_val_score(clf, x, y, cv=5)
scores.mean()
```

```python
def obj_func(solution):
    clf = SVC(kernel='rbf', decision_function_shape='ovo', C=solution[0], gamma=solution[1], random_state=42)
    scores = cross_val_score(clf, x, y, cv=5)
    return -scores.mean()
```

```python
from mealpy.swarm_based.GOA import BaseGOA

lb = [1, 5e-05]
ub = [1000, 0.5]
verbose = True
epoch = 50
pop_size = 10

md1 = BaseGOA(obj_func, lb, ub, verbose, epoch, pop_size)
best_pos1, best_fit1, list_loss1 = md1.train()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
clf = SVC(kernel='rbf', decision_function_shape='ovo', C=3.38435886, gamma=0.32981013)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```
